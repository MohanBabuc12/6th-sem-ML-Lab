```python
# Step 1: Import required libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Downloading Stock Market Data
# Define the ticker symbols for Indian banks
tickers = ["HDFCBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS"]

# Fetch historical data for the year 2024
data = yf.download(tickers, start="2024-01-01", end="2024-12-30", group_by='ticker')

# Display the first 5 rows of the dataset for each ticker
print("First 5 rows of the dataset:")
print(data.head())

# Step 3: Basic Data Exploration
# Check the shape of the dataset
print("\nShape of the dataset:")
print(data.shape)

# Check column names
print("\nColumn names:")
print(data.columns)

# Plot the closing price and daily returns for all three banks
plt.figure(figsize=(12, 12))

# Loop through each ticker and plot the data
for i, ticker in enumerate(tickers, start=1):
    bank_data = data[ticker]

    # Calculate daily returns
    bank_data['Daily Return'] = bank_data['Close'].pct_change()

    # Subplot for closing price
    plt.subplot(3, 2, 2*i-1)
    bank_data['Close'].plot(title=f"{ticker} - Closing Price")

    # Subplot for daily returns
    plt.subplot(3, 2, 2*i)
    bank_data['Daily Return'].plot(title=f"{ticker} - Daily Returns", color='orange')

plt.tight_layout()
plt.show()

# Step 4: Saving the Processed Data to a New CSV File (Optional)
# Save each bank's data to a CSV file
for ticker in tickers:
    bank_data = data[ticker]
    bank_data.to_csv(f'{ticker}_stock_data.csv')
    print(f"\n{ticker} stock data saved to '{ticker}_stock_data.csv'.")
```
